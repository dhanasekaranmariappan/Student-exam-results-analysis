# Student Exam Analysis
## Key Findings:
- 917/1000 students failed at least one subject
- Science has highest failure rate (35.1%)
- English has lowest failure rate (32.8%)
- Passing students score 112 points higher on average
- Domain knowledge matters: fail one subject = fail overall

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("jayantiprasad/student-exam-results-dataset")

print("Path to dataset files:", path)
```

```python
import pandas as pd
#moved to my preferred location
df_one = pd.read_csv('C:\\Users\\DHANASEKARAN\\Documents\\conda\\.ipynb_checkpoints\\results.csv')
print(df_one)
```

```python
print(df_one.isna().sum())
```

```python
columns_one = df_one.iloc[1:1]
subjects = ["Hindi", "English", "Science", "Maths",	"History", "Geograpgy"]
i = 0
for each_subject in subjects:
    if each_subject in columns_one:
        i += 1
print(f"Total number of tests in this term: {i}")
```

```python
six_subject_columns_with_marks = df_one.iloc[: ,1:7]
print(six_subject_columns_with_marks)
df_one['failed'] = (six_subject_columns_with_marks < 35).sum(axis=1)
no_of_failed = (df_one['failed'] > 0).sum()
no_of_passed = (df_one['failed'] == 0).sum()
print('Total failed students: ', no_of_failed)
print('Total passsed students: ', no_of_passed)
```

```python
# Students who failed based on TOTAL score only (total < 210)
total_failures = (df_one['Total'] < 210).sum()
print("Failed by total score:", total_failures)
```

```python
import matplotlib.pyplot as plt

df_one['Total'].hist(bins=30)
plt.title('Distribution of Total Scores')
plt.xlabel('Total Score')
plt.ylabel('Number of Students')
plt.show()
print(df_one['Total'].describe())
```

```python
failed_students = df_one[df_one['failed']>0]
average_failed_score = failed_students['Total'].mean()
print(f'Average total score of failing students: {average_failed_score}')

passed_students = df_one[df_one['failed']==0]
average_passed_score = passed_students['Total'].mean()
print(f'Average total score of passed students: {average_passed_score}')
```

```python
subjects = ['Hindi', 'English', 'Science', 'Maths', 'History', 'Geograpgy']
failure_rate = (df_one[subjects] < 35).mean()
print(failure_rate)
print(f'\nMost failed subject: {failure_rate.idxmax()}')
print(f'Least failed subject: {failure_rate.idxmin()}') 
```

```python
failure_rate.plot(kind='bar', color='tomato')
plt.title('Failure Rate by Subject')
plt.ylabel('Failure Rate')
plt.xticks(rotation = 30)
plt.tight_layout()
plt.show()
```
